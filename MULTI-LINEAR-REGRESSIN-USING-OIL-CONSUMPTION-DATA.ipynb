```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\sunny\Downloads\petrol_consumption.csv")
# Index(['Petrol_tax', 'Average_income', 'Paved_Highways',
#        'Population_Driver_licence(%)', 'Petrol_Consumption'],
#       dtype='object')

x = data[["Petrol_tax","Average_income","Paved_Highways","Population_Driver_licence(%)"]]
y = data[["Petrol_Consumption"]]
x
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size = 0.2,random_state = 0)
print("test data shape ",xtest.shape)
print("train data shape", xtrain.shape)
```

```python
# coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
lnn = LinearRegression()
lnn.fit(xtrain, ytrain)

# how different parameters are affecting the consumption of oil..
coe = pd.DataFrame(lnn.coef_, columns =['Petrol_tax', 'Average_income', 'Paved_Highways',
'Population_Driver_licence(%)'])
coe
```

```python
yp = lnn.predict(xtest)
acc = lnn.score(xtest, ytest)
acc
```
